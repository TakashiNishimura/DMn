```python
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-x**2/5.0)
plt.plot(x,y,'o')
plt.grid(True)
```

```python
from scipy import interpolate
f = interpolate.interp1d(x, y, 'linear')
plt.plot(x,f(x),'-')
plt.grid(True)
```

```python
f2 = interpolate.interp1d(x, y, 'cubic')
xnew = np.linspace(0, 10, num=30, endpoint=True)
plt.plot(xnew, f2(xnew), '--')
plt.grid(True)
```

```python
plt.plot(x, y ,'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.grid(True)
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def lorenz_func(v, t, p, r, b):
 return [-p*v[0]+p*v[1], -v[0]*v[2]+r*v[0]-v[1], v[0]*v[1]-b*v[2]]
```

```python
p = 10
r = 28
b = 8/3
v0 = [0.1, 0.1, 0.1]
t = np.arange(0, 100, 0.02)
v = odeint(lorenz_func, v0, t, args=(p, r ,b))
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(v[:,0], v[:, 1], v[:,2])
plt.title('Lorenz')
plt.grid(True)
```

# 演習5-5.

ローレンツ方程式とは、カオス的ふるまいを示す非線型常微分方程式のこと。  
関数lorenz_funcにある通りv0, v1, v2の方程式で、  
dv0/dt = -pv0 + pv1, dv1/dt = -v0v2 + rv0 -v1, dv2/dt = v0v1 -bv2  
のうち、p, r, bの3つの定数によってそのふるまいが決まる。  
また、カオスとは、わずかに初期条件が変わるだけで結果に大きな差が起こる現象や、  
予想がつかないような複雑な様子を示す現象を扱う理論のこと。  
積分法による解が得られないうえ、計算機においても数値解析過程で生じる誤差によって  
正確な値を得ることが事実上不可能である。
