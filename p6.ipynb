```python
data1 = {
  'id': ['100','102','103'],
  'name': ['Taro', 'Jiro', 'Yuzuru'],
  'city': ['Tokyo', 'Osaka', 'Sendai'],
  'birth_year': [1993, 2001, 1994]
}
df1 = DataFrame(data1)
df1
```

```python
data2 = {
  'id': ['100', '101', '103'],
  'math': [100, 70, 40],
  'english': [60, 80, 100]
}
df2 = DataFrame(data2)
df2
```

```python
print('(内部)結合')
pd.merge(df1,df2, on = 'id')
```

```python
print('左外部結合')
pd.merge(df1,df2, how = 'left')
```

```python
print('右外部結合')
pd.merge(df1,df2, how = 'right')
```

```python
print('全結合')
pd.merge(df1,df2, how = 'outer')
```

```python
data3 = {
  'id': ['100', '101', '103', '105', '106', '107', '108', '103'],
  'name': ['Taro', 'Mai', 'Yuzuru', 'Syoma', 'Satoko', 'Rika', 'Kaori', 'Yuzuru'],
  'city': ['Tokyo', 'Kobe', 'Sendai', 'Nagoya', 'Osaka', 'Osaka', 'Kobe', 'Sendai'],
  'birth_year': [1993, 2001, 1994, 1999, 1998, 2004, 2003, 1994]
}
df3 = DataFrame(data3)
df3
```

```python
df13 = pd.concat([df1,df3])
df13
```

```python
city_map ={
  'Sendai': 'Tohoku',
  'Tokyo': 'Kanto',
  'Nagoya': 'Chubu',
  'Osaka': 'Kansai',
  'Kobe': 'Kansai'
}
city_map
```

```python
df13['region'] = df13['city'].map(city_map)
df13
```

```python
df13['decades'] =df13['birth_year'].map(lambda x: str(x)[0:3])
df13
```

```python
birth_year_bins = [1990, 1995, 2000, 2005]
birth_year_cut_data = pd.cut(df13.birth_year, birth_year_bins)
birth_year_cut_data
```

```python
birth_year_cut_data = pd.cut(df13.birth_year, birth_year_bins, right=False)
birth_year_cut_data
```

```python
pd.value_counts(birth_year_cut_data)
```

```python
group_names = ['early1990s', 'late1990s', 'early2000s']
birth_year_cut_data = pd.cut(df13.birth_year, birth_year_bins, labels = group_names)
pd.value_counts(birth_year_cut_data)
```

```python
df13 = df13.drop_duplicates()
df13
```

```python
df13.groupby(['region','city'], as_index = False)['birth_year'].mean()
```

```python
for group, subdf in df13.groupby('region'):
  print('==========================================================')
  print('Region Name:{0}'.format(group))
  print(subdf)
```
