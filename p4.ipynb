# 問題1.
変動係数とは、標準偏差を相加平均で割ったもの。  
単位の異なるデータ、平均値に対するデータとばらつきの関係を相対的に評価する。
  
上の結果では、'failures'の散らばりが大きい。

```python
plt.plot(sdata_m['G1'], sdata_m['G3'], 'o')
plt.ylabel('G3 grade')
plt.xlabel('G1 grade')
plt.grid(True)
```

```python
print('共分散行列',np.cov(sdata_m['G1'],sdata_m['G3']))
print('G1の分散',sdata_m['G1'].var())
print('G3の分散',sdata_m['G3'].var())
```

```python
sp.stats.pearsonr(sdata_m['G1'],sdata_m['G3'])
```

```python
np.corrcoef([sdata_m['G1'],sdata_m['G3']])
```

```python
sns.pairplot(sdata_m[['Dalc','Walc','G1','G3']])
plt.grid(True)
```

```python
plt.plot(sdata_m['G1'], sdata_m['G3'], 'o')
plt.ylabel('G3 grade')
plt.xlabel('G1 grade')
plt.grid(True)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
X = sdata_m.loc[:, ['G1']].values
Y = sdata_m['G3'].values
reg.fit(X, Y)
print('回帰係数：', reg.coef_)
print('切片：', reg.intercept_)
```

```python
plt.scatter(X,Y)
plt.ylabel('G3 grade')
plt.xlabel('G1 grade')
plt.plot(X,reg.predict(X))
plt.grid(True)
```

```python
print('決定係数：', reg.score(X,Y))
```

# 問題2.
決定係数とは、回帰分析によって求められた目的変数の予測値が、  
実際の目的変数の値とどのくらい一致しているかを表す指標。  
1に近いほど、回帰式が実際のデータに当てはまっていると言える。
