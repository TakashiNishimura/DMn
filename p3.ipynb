```python
t_data3 = {'ID':['100','200','300','400'],
           'Class':['A','B','B','A'],
           'Math':[100,49,78,87],
           'Eng':[43,80,83,90]}
f_data3 = DataFrame(t_data3)
f_data3
```

```python
pd.merge(f_data1, f_data3)
```

```python
print(f_data3.groupby('Class')['Math'].mean())
print(f_data3.groupby('Class')['Math'].max())
```

```python
f_data3.Math.sort_values()
```

```python
f_data3.sort_values('Math')
```

```python
f_data3.sort_values('Math',ascending=False)
```

```python
import numpy as np
f_data3 = f_data3.replace('B',np.nan)
f_data3
```

```python
f_data3.isnull()
```

```python
f_data3.isnull().sum()
```
