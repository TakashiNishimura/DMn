```python
import scipy.linalg as linalg
matrix = np.array([[1,-1,-1],[-1,1,-1],[-1,-1,1]])
print('行列 A=')
print(matrix)
print('行列式 detA=',linalg.det(matrix))
print('逆行列 A^{-1}=')
print(linalg.inv(matrix))
```

```python
eig_value, eig_vector = linalg.eig(matrix)
print('固有値=',eig_value)
print('固有ベクトル=')
print(eig_vector)
```

```python
#演習2
matrix2 = np.array([[1,2,3],[2,1,3],[3,-1, 2]])
print('行列 A=')
print(matrix2)
print('行列式 detA=',linalg.det(matrix2))
print('逆行列 A^{-1}=')
print(linalg.inv(matrix))
eig_value2, eig_vector2 = linalg.eig(matrix2)
print('固有値=',eig_value2)
print('固有ベクトル=')
print(eig_vector2)
```

```python
def func(x):
 return (x**2 + 2*x + 1)
x = np.arange(-10,10)
plt.figure(figsize = (20,6))
plt.plot(x, func(x))
plt.grid(True)
```

```python
from scipy.optimize import newton
# 引数はfunx(x)=0，解－1 の近似値が求まる
print(newton(func,0))
```

```python
#演習3
def func2(x):
    return (x**3 + 3*(x**2) + 4*x + 2)
x = np.arange(-10,10)
plt.figure(figsize = (20,6))
plt.plot(x, func2(x))
plt.grid(True)

print(newton(func2, 0))
```
